```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]

plt.figure(figsize=(5,3), dpi=100)

plt.plot(x, y, label = '2x', color = 'brown',linestyle='--',linewidth=2,
         marker='.', markersize=10, markeredgecolor='yellow')

x2 = np.arange(0, 4.5, 0.5)
plt.plot(x2, x2**2)

plt.title('Our First Graph!')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.legend()
plt.show()
```

```python
labels =['A','B','C']
values = [1,4,2]

plt.bar(labels, values)
plt.figure(figsize = (6, 4))
plt.show()
```
