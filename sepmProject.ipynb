```python
def convert_sqft_to_num(x):
    tokens = x.split('-')
    if len(tokens) == 2:
        return (float(tokens[0])+float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
        #key holds the name of the location (the unique value in the location column).
        #subdf is a DataFrame containing all rows from df that belong to that specific location.
def remove_pps_outliers(df):
    df_out = pd.DataFrame() #create empty dataframe
    for key, subdf in df.groupby('location'):
      m = np.mean(subdf.price_per_sqft)     #mean of pps within group(location)
      st = np.std(subdf.price_per_sqft)    #std dev of pps within group(location)
      reduced_df = subdf[(subdf.price_per_sqft>(m-st)) & (subdf.price_per_sqft<=(m+st))]
      df_out = pd.concat([df_out,reduced_df],ignore_index=True)
    return df_out
df7 = remove_pps_outliers(df6)
df7.shape
```

This function removes properties with an unusually high or low price per square foot within each location, leaving only those properties whose price per square foot is close to the average for that location. The result is a cleaner dataset that’s less likely to be skewed by extreme values.


```python
def plot_scatter_chart(df,location):
    bhk2 = df[(df.location==location) & (df.bhk==2)] #select rows where location matches location and bhk == 2
    bhk3 = df[(df.location==location) & (df.bhk==3)]
    matplotlib.rcParams['figure.figsize'] = (15,10) #(length, breadth) in inches
    plt.scatter(bhk2.total_sqft,bhk2.price,color='blue',label='2 BHK', s=50)
    plt.scatter(bhk3.total_sqft,bhk3.price,marker='+', color='green',label='3 BHK', s=50)
    plt.xlabel("Total Square Feet Area")
    plt.ylabel("Price (Lakh Indian Rupees)")
    plt.title(location)
    plt.legend()

plot_scatter_chart(df7,"Whitefield")
```

```python
def remove_bhk_outliers(df):
    exclude_indices = np.array([]) #index of rows which are outliers
    for location, location_df in df.groupby('location'):
      bhk_stats = {}
      for bhk, bhk_df in location_df.groupby('bhk'):
          bhk_stats[bhk] = {
              'mean': np.mean(bhk_df.price_per_sqft),
              'std': np.std(bhk_df.price_per_sqft),
              'count': bhk_df.shape[0]
          }
      for bhk, bhk_df in location_df.groupby('bhk'):
          stats = bhk_stats.get(bhk-1)
          if stats and stats['count']>5:
              exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
    return df.drop(exclude_indices,axis='index')

df8 = remove_bhk_outliers(df7)
df8.shape
#to remove rows in which price of bhk < price of bhk - 1 room
```

# **MODEL BUILDING (VIDEO-4)**

```python
dummies = pd.get_dummies(df10.location)
dummies.head()
```

```python
df11 = pd.concat([df10,dummies.drop('other',axis='columns')],axis='columns')
df11.head()
```

```python
df12 = df11.drop('location', axis = 'columns')
df12.head()
```

```python
df12.shape
```

```python
x = df12.drop('price',axis='columns')
x.head()
```

```python
y = df12.price
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr_clf = LinearRegression()
lr_clf.fit(x_train,y_train)
lr_clf.score(x_test,y_test)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)

cross_val_score = cross_val_score(LinearRegression(), x, y, cv=cv)
cross_val_score
```

```python
from sklearn.model_selection import GridSearchCV, ShuffleSplit
from sklearn.linear_model import Lasso, LinearRegression
from sklearn.tree import DecisionTreeRegressor
import pandas as pd

def find_best_model_using_gridsearchcv(x, y):
    algos = {
        'linear_regression': {
            'model': LinearRegression(),
            'params': {
                'fit_intercept': [True, False]
            }
        },
        'lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [1, 2],
                'selection': ['random', 'cyclic']
            }
        },
        'DecisionTreeRegressor': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion': ['squared_error', 'friedman_mse'],
                'splitter': ['best', 'random']
            }
        }
    }

    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)

    for algo_name, config in algos.items():
        gs = GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(x, y)
        scores.append({
            'model': algo_name,
            'best score': gs.best_score_,
            'best_params': gs.best_params_
        })

    return pd.DataFrame(scores, columns=['model', 'best score', 'best_params'])

find_best_model_using_gridsearchcv(x, y)
```

```python
def predict_price(location,sqft,bath,bhk):
    loc_index = np.where(x.columns==location)[0][0]

    X = np.zeros(len(x.columns))
    X[0] = sqft
    X[1] = bath
    X[2] = bhk
    if loc_index >= 0:
        X[loc_index] = 1
    return lr_clf.predict([X])[0]
```
